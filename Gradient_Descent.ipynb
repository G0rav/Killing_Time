```
import numpy as np
import matplotlib.pyplot as plt
 
#plt.rcParams['figure.figcolor'] = 'white'
```

```
x = np.arange(-10,10,.1)
 
def y(x):
  return ( x**6 + x**5 - x**4 - 10*x**3 + 8*x + 2)
def dy(x):
  return ( 6*x**5 + 5*x**4 - 4*x**3 - 30*x**2 + 8)
```

```
plt.plot(x,y(x))
plt.plot(x,dy(x), label ='dy' )
plt.legend()
plt.show()
```

```
x_hat = 4
lr = .001
 
plt.plot(x,y(x))
for i in range(5):
  print(i,'-',x_hat)
  plt.scatter(x_hat,y(x_hat), color='red')
  z = dy(x_hat)
 
  x_hat -= lr*z
 
plt.show()
```

```
def f(x):
  return (x*np.sin(x) + .5* x**2)
 
def df(x):
  return (x*np.cos(x) + np.sin(x) + x)
```

```
plt.plot(x,f(x))
plt.plot(x,df(x), label= 'df')
plt.axhline(0)
plt.axvline(0)
plt.legend()
plt.show()
```

```
x_hat = 8
lr = .1
 
plt.plot(x,f(x))
for i in range(50):
  print(i,'-',x_hat)
  plt.scatter(x_hat,f(x_hat), color='red')
  z = df(x_hat)
 
  x_hat -= lr*z
 
plt.show()
```

```
def sigmoid(x):
  return (1/(1+np.exp(-x)))
 
def ds(x):
  return sigmoid(x)*(1-sigmoid(x))
```

```
x = np.linspace(-10,10,100)
 
plt.plot(x,sigmoid(x))
plt.plot(x,ds(x), label = 'ds')
plt.legend()
plt.show()
```
